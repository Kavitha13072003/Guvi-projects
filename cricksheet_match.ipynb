```python
import pandas as pd
import json
import os


folder_path = '/content/'


json_files = [file for file in os.listdir(folder_path) if file.endswith('.json')]


print(f"Total JSON files found: {len(json_files)}")

```

```python

with open(folder_path + json_files[0]) as f:
    data = json.load(f)


print(data['info'])

```

```python

match_data = []

for file in json_files:
    with open(folder_path + file) as f:
        data = json.load(f)


        match_info = {
            'match_id': file,
            'team1': data['info']['teams'][0],
            'team2': data['info']['teams'][1],
            'venue': data['info'].get('venue', 'Unknown'),
            'city': data['info'].get('city', 'Unknown'),
            'date': data['info']['dates'][0],
            'match_type': data['info'].get('match_type', 'Unknown'),
            'toss_winner': data['info']['toss']['winner'],
            'toss_decision': data['info']['toss']['decision'],
            'winner': data['info'].get('outcome', {}).get('winner', 'Draw')
        }

        match_data.append(match_info)


df_matches = pd.DataFrame(match_data)

df_matches.head()

```

```python
df_matches.to_csv('/content/match_summary.csv', index=False)
```

```python
deliveries_data = []

for file in json_files:
    with open(folder_path + file) as f:
        data = json.load(f)
        match_id = file

        # Check innings exists
        if 'innings' in data and isinstance(data['innings'], list):
            for innings in data['innings']:
                team = innings.get('team', 'Unknown')
                overs = innings.get('overs', [])

                for over in overs:
                    over_number = over.get('over')
                    deliveries = over.get('deliveries', [])

                    for delivery in deliveries:
                        batter = delivery.get('batter', 'Unknown')
                        bowler = delivery.get('bowler', 'Unknown')
                        runs = delivery.get('runs', {}).get('total', 0)
                        wicket_info = delivery.get('wickets', [])
                        wicket_type = wicket_info[0]['kind'] if wicket_info else None

                        deliveries_data.append({
                            'match_id': match_id,
                            'team': team,
                            'over': over_number,
                            'batsman': batter,
                            'bowler': bowler,
                            'runs': runs,
                            'wicket_type': wicket_type
                        })


import pandas as pd
df_deliveries = pd.DataFrame(deliveries_data)


df_deliveries.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Top 5 batsmen
top_batsmen = df_deliveries.groupby('batsman')['runs'].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(8,5))
sns.barplot(x=top_batsmen.values, y=top_batsmen.index, palette='viridis')
plt.title('Top 5 Batsmen by Total Runs')
plt.xlabel('Runs')
plt.ylabel('Batsman')
plt.show()
```

```python

dismissals = df_deliveries[df_deliveries['wicket_type'].notnull()]

top_bowlers = dismissals['bowler'].value_counts().head(5)

plt.figure(figsize=(8,5))
sns.barplot(x=top_bowlers.values, y=top_bowlers.index, palette='magma')
plt.title('Top 5 Bowlers by Number of Wickets')
plt.xlabel('Wickets')
plt.ylabel('Bowler')
plt.show()
```

```python
wicket_counts = df_deliveries['wicket_type'].value_counts()

plt.figure(figsize=(7,5))
wicket_counts.plot(kind='bar', color='salmon')
plt.title('Count of Wicket Types')
plt.xlabel('Dismissal Type')
plt.ylabel('Number of Wickets')
plt.show()
```

```python
team_runs = df_deliveries.groupby('team')['runs'].sum().sort_values(ascending=False)

plt.figure(figsize=(9,6))
sns.barplot(x=team_runs.values, y=team_runs.index, palette='cool')
plt.title('Total Runs by Team')
plt.xlabel('Total Runs')
plt.ylabel('Team')
plt.show()
```

```python
over_runs = df_deliveries.groupby('over')['runs'].sum()

plt.figure(figsize=(10,5))
over_runs.plot(marker='o', color='green')
plt.title('Total Runs Scored in Each Over')
plt.xlabel('Over Number')
plt.ylabel('Runs')
plt.grid(True)
plt.show()
```
